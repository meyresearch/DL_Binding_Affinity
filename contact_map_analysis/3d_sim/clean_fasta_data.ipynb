```python
import numpy as np
```

```python
import glob
import os
```

```python
files = glob.glob('../data_main/fasta_files/*.fasta')
```

```python
out_file = open('fasta_lengths.txt', 'w')
for f in files:
    fh = open(f)
    data = fh.readlines()
    name = os.path.basename(f)
    if len(data[-1]) < 50:
        continue
    out_file.write(str(len(data[-1]))+','+name.strip('.fasta')+'\n')
    print(len(data[-1]))
out_file.close()
```
